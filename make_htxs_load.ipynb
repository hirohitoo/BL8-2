```python
HTXS_LOAD_TIMESTAMP="20260415"
SAMPLE_FILE="Sample-spreadsheet-04142026a.csv"
X_POS=[0]
Z_POS=[-0.5,0,0.5]
PRI=["xas_ov500","xas_c1s","xas_o1s","xps_ov","xps_c1s","xps_o1s"]
AG_DICT={"xas_ov500":[90],"xas_c1s":[90],"xas_o1s":[90],"xps_c1s":[20],"xps_ov":[20], "xps_o1s":[20]}
ME_DICT={"xas_ov500":"xas_ov500","xas_c1s":"xas_c1s","xas_o1s":"xas_o1s","xps_c1s":"xps_c1s","xps_ov":"xps_ov", "xps_o1s":"xps_o1s"}
BL_DICT={"xas_ov500":"xas_ov500_prep","xas_c1s":"xas_c1s_blprep","xas_o1s":"xas_o1s_blprep","xps_c1s":"xps_blprep","xps_ov":"xps_blprep", "xps_o1s":"xps_blprep"}

with open("htxs_load.mac", "w") as f:
    print("running htxs_load.mac...",file=f)
    print("HTXS_LOAD_ID=1",file=f)
    print("HTXS_LOAD_TIMESTAMP="+HTXS_LOAD_TIMESTAMP,file=f)
    print("def htxs_set_stick_geometry_override '{",file=f)
    print("    HTXS_STICK_STARTINC=90",file=f)
    print("    HTXS_STICK_ENDINC=260",file=f)
    print("}'",file=f)
#reset the sample stick info, and load it fresh from below
    print("HTXS_SAMPLE_ID=0",file=f)
# Samples
    
    import csv
    import ast
    with open(SAMPLE_FILE, "r") as g:
        reader = csv.reader(g)
        measure_list=[]
        sample_dict_str='{'
        for row in reader:
            print("htxs_set_sample_info",row[1],row[2],'"'+row[4]+'"',row[7],row[8],row[9],row[11],row[12],row[13],file=f)  # Each row is a list of strings
            measure_list.extend(list(row[27][1:-1].split(',')))
            sample_dict_str=sample_dict_str+'"'+str(int(row[1])*100+int(row[2]))+'":'+'"'+row[4]+'",'
        measure_list=list(set(measure_list))
        sample_dict_str=sample_dict_str+'}'
    SM_DICT=ast.literal_eval(sample_dict_str)
with open("htxs_queue.mac", "w") as f:    
    print("def peakcnton 'PEAK_CNTRL_ENABLE_CNT=1'",file=f)
    print("def peakcntoff 'PEAK_CNTRL_ENABLE_CNT=0'",file=f)
    print("def setaugerBE(be) '{peakcntoff;xps_be("+'"SetAugerBE"'+",A[mono],be-2, be, 0.1, 0.19);}'",file=f)
    print("htxs_refresh",file=f)
    print("qdo htxs_load.mac",file=f)
    # Sort data based on the index in the priority list
    sorted_measure_list = sorted(measure_list, key=lambda x: PRI.index(x))
    # Result: ['high', 'medium', 'low', 'low']
    for measure in sorted_measure_list:
         with open(SAMPLE_FILE, "r") as g:
            reader = csv.reader(g)
            for row in reader:

                if measure in row[27]:
                    SM_NUM=str(int(row[1])*100+int(row[2]))
                    print("#",measure,SM_DICT[SM_NUM],file=f)
                    for _x in X_POS:
                        for _z in Z_POS:
                            for _ag in AG_DICT[measure]:
                                print('htxs_set_pos_samplecenter('+SM_NUM+','+str(180)+"-"+str(_ag)+str(_x)+',0,'+str(_z)+')'+';htxs_move;'+BL_DICT[measure]+';htxs_mopt;htxs_pesopt;'+ME_DICT[measure],file=f)
```
